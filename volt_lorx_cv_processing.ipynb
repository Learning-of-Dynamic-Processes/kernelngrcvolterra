```python
import numpy as np
import matplotlib.pyplot as plt
from utils.cleaning import param_sort, err_sort
%matplotlib widget
```

```python
mses_param = param_sort("crossvalresults/cv_lorenz_x_volterra_meansquare.txt")
wass_param = param_sort("crossvalresults/cv_lorenz_x_volterra_wasserstein.txt")
specdens_param= param_sort("crossvalresults/cv_lorenz_x_volterra_specdens.txt")
```

```python
mses_err = err_sort("crossvalresults/cv_lorenz_x_volterra_meansquare.txt")
wass_err = err_sort("crossvalresults/cv_lorenz_x_volterra_wasserstein.txt")
specdens_err= err_sort("crossvalresults/cv_lorenz_x_volterra_specdens.txt")
```

```python
mses = []
wass = []
specdens = []
for param_id in range(len(mses_param)):
    mses.append(mses_param[param_id][1])
    wass.append(wass_param[param_id][1])
    specdens.append(specdens_param[param_id][1])
```

```python
mses_low = []
wass_low = []
specdens_low = []
param_low = []
for param_id in range(len(mses_param)):
    mse = mses_param[param_id][1]
    wass = wass_param[param_id][1]
    specdens = specdens_param[param_id][1]
    if specdens <= 6e-3 and wass <= 10:
        mses_low.append(mse)
        wass_low.append(wass)
        specdens_low.append(specdens)
        param_low.append(param_id)
```
